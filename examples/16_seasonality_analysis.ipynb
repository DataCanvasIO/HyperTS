````python
from hyperts.toolbox import TSToolBox
from hyperts.toolbox import plot_seasonality
from hyperts.datasets import load_network_traffic
````

````python
df_traffic = load_network_traffic()
df_traffic.head()
````

````python
df_traffic.info()
````

````python
tb = TSToolBox()
````

````python
df_traffic['TimeStamp'] = tb.to_datetime(df_traffic['TimeStamp'])
````

````python
plot_seasonality(df_traffic, timestamp='TimeStamp', show_ts_col='Var_1')
````

````python
plot_seasonality(df_traffic, timestamp='TimeStamp', show_ts_col='Var_2')
````

````python
plot_seasonality(df_traffic, timestamp='TimeStamp', show_ts_col='Var_6')
````

It is noteworthy that from a weekly perspective, the network traffic usage on ```Wednesday``` is underestimated and gradually increases, reaching its peak on ```Saturday``` (due to the weekend break). From a daily perspective, the traffic usage is low during the early morning hours (while people are sleeping), gradually increasing during the day, and reaching a peak during the hours of **8pm to 11pm** in the evening.
